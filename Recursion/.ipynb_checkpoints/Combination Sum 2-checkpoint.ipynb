# Combination Sum 2 

Given a collection of candidate numbers (candidates) and a target number (target), find all unique combinations in candidates where the candidate numbers sum to target. Each number in candidates may only be used once in the combination.

Note: The solution set must not contain duplicate combinations.

Example 1:

Input: candidates = [10,1,2,7,6,1,5], target = 8

Output: 
[
[1,1,6],
[1,2,5],
[1,7],
[2,6]]


Explanation: These are the unique combinations whose sum is equal to target.
 
Example 2:

Input: candidates = [2,5,2,1,2], target = 5

Output: [[1,2,2],[5]]

Explanation: These are the unique combinations whose sum is equal to target.
    

```python
def Combination(index,target,array,list1,result):
    if target==0:
        result.append(list1[:])
        return
    # Take Logic The solution set must not contain duplicate combinations.
    for i in range(index,len(array)):
        if i>index and array[i]==array[i-1]:
            continue
        if array[i]>target:
            break
    
        list1.append(array[i])
        Combination(i+1,target-array[i],array,list1,result)  # Take call
    
    # Backtracking
        list1.pop()
    
    # Not Take logic 
        #Combination(index+1,target,array,list1)  # Not take
    
array = [2, 5, 2, 1, 2]
array.sort()  # Sorting to handle duplicates
target = 5
result = []
Combination(0, target, array, [], result)
print(result)

array2 = [10, 1, 2, 7, 6, 1, 5]
array2.sort()
target2 = 8
result2 = []
Combination(0, target2, array2, [], result2)
print(result2)
```
