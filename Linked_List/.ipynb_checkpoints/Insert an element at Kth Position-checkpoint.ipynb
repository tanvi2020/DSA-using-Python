# Insert an element at Kth Position

- Create a new box (the new node).

- Make it point to the next box.

- And tell the current box to point to this new one.

new node next -> current.next

current.next -> new node


```python
# Insert an element at Kth Position
class Node:
    def __init__(self,value):
        self.data=value
        self.next=None
        
def create_linked_list(values):
    head=Node(values[0])
    mover=head
    for v in values[1:]:
        mover.next=Node(v)
        mover=mover.next
    return head

def insert_at_kth_position(head,k,val): 
    if k==1:
        new_node=Node(val)
        new_node.next=head
        return new_node
    
    counter=1
    current=head
    while current is not None:
        if counter==k-1:
            new_node=Node(val) 
            new_node.next=current.next
            current.next=new_node
            return head
        current = current.next
        counter+=1
    return head

def print_list(head):
    current=head
    while current:
        print(current.data,end='->')
        current=current.next
    print('None')

values = [1, 2, 3, 4, 5]
head = create_linked_list(values)
i=insert_at_kth_position(head,3,100)
print_list(i)    
```
